```python
def plotting(df,key):
    df_=df[df['cloth_type']==key]
    url_=df_['thumbnail_url'][15:30] #[:15], [45:60], etc...
    data_=[i for i in url_]
    f=[io.imread(i) for i in data_]
    #####
    for i,j in enumerate(f):
        plt.figure(figsize=(10,10))
        plt.subplot(3,5,i+1)
        plt.imshow(j)
        plt.title(key)
    
```

```python
# Get images from url_link and save them to corresponding folder:
import urllib.request
def save_img(data):
    cnt=0
    folder_size=0
    location=[]
    for i in range(data.shape[0]):
        cnt+=1
        url=data.iloc[i][0]
        folder=data.iloc[i][1]
        local='IMG/'+str(folder)+'/'+str(folder)+'_'+str(cnt)+'.jpg'
        location.append(local)
        urllib.request.urlretrieve(url,local)
```
